# 1. Distance from a Line

<br>

$$\omega = \begin{bmatrix}\omega_1 \\ \omega_2\end{bmatrix}, \, x = \begin{bmatrix}x_1\\x_2\end{bmatrix} \; \implies g(x) = \omega_0 + \omega^Tx = \omega_0 + \omega_1x_1 + \omega_2x_2 $$



<br><br>
<center></center>
<br>

(1) If $\vec p$ and $\vec q$ are on the decision line

<br>

$$
\begin{align*}
g\left(\vec p\right) = g\left(\vec q\right) = 0 \quad
& \implies \quad \omega_0 + \omega^T \vec p = \omega_0 + \omega^T \vec q = 0 \\
& \implies \quad \omega^T \left( \vec p- \vec q \right) = 0
\end{align*}
$$

<br>

$$
\begin{align*}
& \therefore \, \omega : \text{orthogonal to the line}  \\
& \implies \text{tells the direction of the line}
\end{align*}$$

<br>

(2) Compute $d$ which is a normal signed distance from the origin to the line

<br>

- If $x$ is on the line and $x = d\frac{\omega}{\lVert \omega \rVert}$ (where $d$ is a normal distance from the origin to the line)

<br>

$$ \begin{align*}  g(x)& = \omega_0 + \omega^Tx = 0 \;  \\
& \Rightarrow \omega_0 + \omega^Td\frac{\omega}{\lVert \omega \rVert} = \omega_0 + d\frac{\omega^T\omega}{\lVert \omega \rVert} = \omega_0 + d\lVert \omega \rVert = 0 \\\\
& \therefore d \, = - \frac{\omega_0}{\lVert \omega \rVert}
\end{align*}$$

<br>

(3) Compute $h$ which is a normal signed distance from $x$ to the line

<br>

- For any vector of $x$

<br>

$$ x = x_{\perp} + r \frac{\omega}{\lVert \omega \rVert}$$

<br>

$$ \omega^Tx = \omega^T \left( x_{\perp} + r \frac{\omega}{\lVert \omega \rVert}\right) = r \frac{\omega^T\omega}{\lVert \omega \rVert} = r \lVert \omega \rVert$$

<br>

$$ \begin{align*} g(x)
& = \omega_0 + \omega^Tx \\
& = \omega_0 + r \lVert \omega \rVert \qquad (r = d + h) \\
&  = \omega_0 + (d +h) \lVert \omega \rVert\\
& = \omega_0 + \left(- \frac{\omega_0}{\lVert \omega \rVert} + h \right)\lVert \omega \rVert\\
&  = h \lVert \omega \rVert
\end{align*}$$

<br>

$$\therefore \; h = \frac{g(x)}{\lVert \omega \rVert} \implies\; \text{orthogonal signed distance from the line}$$

<br>

- $\text{sign}(h)$ indicates whether $x$ is located on the same side of the hyperplane relative to $\omega$ or not.

# 2. Distance Between Two Parallel Lines $g(x) = -1$ and $g(x) = 1$

(1) Find a distance between $g(x) = -1$ and $g(x) = 1$

- Suppose $g(x_1) = -1,\; g(x_2) = 1$

<br>

$$
\begin{align*}
h_1 &= \frac{g(x_1)}{\lVert \omega \rVert} = \frac{-1}{\lVert \omega \rVert} \\
h_2 &= \frac{g(x_2)}{\lVert \omega \rVert} = \frac{1}{\lVert \omega \rVert} \\\\
\lvert \; \lvert h_2 \rvert + \lvert h_1 \rvert \; \rvert &= \frac{1}{\lVert \omega \rVert} + \frac{1}{\lVert \omega \rVert} = \frac{2}{\lVert \omega \rVert}
\end{align*}
$$

<br>

- This distance formula is especially important in SVMs, where maximizing the margin between parallel decision boundaries is key to optimal classification.

<br>

(2) Another method to find a distance between $g(x) = -1$ and $g(x) = 1$

<br>

<br>
<center></center>
<br>

- Suppose $g(x_1) = -1,\; g(x_2) = 1$

<br>

$$ \begin{array}{I}
\omega_0 + \omega^Tx_1 = -1\\
\omega_0 + \omega^Tx_2 = 1\\
\end{array}
\; \implies \; \omega^T(x_2 - x_1) = 2$$

<br>

$$s = \langle\frac{\omega}{\lVert \omega \rVert}, \; x_2 - x_1 \rangle = \frac{1}{\lVert \omega \rVert}\omega^T(x_2 - x_1) = \frac {2}{\lVert \omega \rVert}$$

#  3. CVXPY를 이용한 SVM 예제

- 이진분류(Binary classification)
  - $Class1 -> C1$ and $Class0 -> C0$

<br>

- 입력(Inputs or Features)
<br>
$$x = \begin{bmatrix}x_1 \\ x_2 \end{bmatrix}$$

<br>

- 분류경계선(hyperplane)
<br>

$$
\begin{align*}
\omega_0 + \omega_1x_1 + \omega_2x_2 &= 0 \\\\
\omega_0 + \begin{bmatrix}\omega_1 & \omega_2 \end{bmatrix}
\begin{bmatrix}x_1 \\ x_2 \end{bmatrix} &= 0\\\\
\omega_0 + \omega^Tx &= 0
\end{align*}
$$

<br>


<br>

- 분류경계선(Decision boundary)
<br>

$$\omega_0 + \omega^Tx = 0$$

<br>

- 데이터의 Class 결정:
<br>

$$ \omega_0 + \omega^Tx > 0 \implies x \; \text{는 Class1 :} \; C_1$$

$$ \omega_0 + \omega^Tx < 0 \implies x \; \text{는 Class0 :} \; C_0$$
<br>

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
## training data gerneration

np.random.seed(1)

x1 = 9*np.random.rand(100, 1)
x2 = 6*np.random.rand(100, 1)

g = 1/3*(2*x1 + 3*x2 - 18)

C1 = np.where(g >= 1)[0]
C0 = np.where(g < -1)[0]
```

```python
xp = np.linspace(0,9,100).reshape(-1,1)

# 분류경계선
ypt = -2/3*xp + 6

plt.figure(figsize = (6, 4))
plt.plot(x1[C1], x2[C1], 'go', label = 'Class1')
plt.plot(x1[C0], x2[C0], 'bo', label = 'Class0')
plt.plot(xp, ypt, 'r', linewidth = 3, label = 'Decision boundary')
plt.title('Linearly and Strictly Separable Classes')
plt.xlabel(r'$x_1$')
plt.ylabel(r'$x_2$')
plt.legend(loc = 1)
plt.xlim([0, 9])
plt.ylim([0, 6])
plt.show()
```


- Class 사이의 마진(margin)을 최대로 하는 조건 ($b$ 는 마진 거리)

    1)   For all points in Class 1   $$\omega_0 + \omega^Tx > b$$
    2)   For all points in Class 0   $$\omega_0 + \omega^Tx < b$$

<br>

$$
\begin{align*}
&\omega_0 + \omega^Tx > b, \quad (b>0)\\ \\
\Longleftrightarrow \;&\frac{\omega_0}{b} + \frac{\omega^T}{b}x > 1 \\\\
\Longleftrightarrow \;&\omega'_0 + \omega'^Tx > 1
\end{align*}$$

<br>

```python
# see how data are generated

xp = np.linspace(0,9,100).reshape(-1,1)
ypt = -2/3*xp + 6
plt.figure(figsize = (6, 4))
plt.plot(x1[C1], x2[C1], 'go', label = 'Class1')
plt.plot(x1[C0], x2[C0], 'bo', label = 'Class0')
plt.plot(xp, ypt, 'r', linewidth = 3, label = 'D.B.')
plt.plot(xp, ypt-1, '--k', label = 'Margin')
plt.plot(xp, ypt+1, '--k')
plt.title('Linearly and Strictly Separable Classes')
plt.xlabel(r'$x_1$')
plt.ylabel(r'$x_2$')
plt.legend(loc = 1)
plt.xlim([0, 9])
plt.ylim([0, 6])
plt.show()
```

## 3.1. 첫번째 최적화(Optimization)

<br>

__Settings:__
<br>

- $n \;(=2)$ features

- $m = N + M$ data points in training set
  
  - $N$ belongs to $C_1$ in training set
  - $M$ belongs to $C_0$ in training set

<br>

$$ x^{(i)} = \begin{bmatrix} 1 \\ x^{(i)}_1 \\ x^{(i)}_2 \end{bmatrix}  \;\; \text{with}\; \omega = \begin{bmatrix} \omega_0 \\ \omega_1 \\ \omega_2 \end{bmatrix}\qquad \text{or} \qquad x^{(i)} = \begin{bmatrix} x^{(i)}_1 \\ x^{(i)}_2 \end{bmatrix} \;\text{with}\; \omega = \begin{bmatrix} \omega_1 \\ \omega_2 \end{bmatrix}$$

<br>

- $\omega$ and $\omega_0$ are the unknown variables

__제약조건(Constrains)__

$$
\begin{align*}
C_1-> X_1 = \begin{cases}
\omega_0 + \omega^Tx^{(1)} \geq1\\
\omega_0 + \omega^Tx^{(2)} \geq1\\
 \quad  \quad \vdots \\
\omega_0 + \omega^Tx^{(N)}\geq1\\
\end{cases} \qquad\text{or} \qquad
C_1-> X_1 = \begin{cases}
\omega^Tx^{(1)} \geq1\\
\omega^Tx^{(2)} \geq1\\
 \quad  \quad \vdots \\
\omega^Tx^{(N)} \geq1\\
\end{cases}
\end{align*}
$$
<br>

$$
\begin{align*}
C_0-> X_0 = \begin{cases}
\omega_0 + \omega^Tx^{(N+1)} \leq{-1}\\
\omega_0 + \omega^Tx^{(N+2)} \leq{-1}\\
 \quad  \quad \vdots \\
\omega_0 + \omega^Tx^{(N+M)} \leq{-1}\\
\end{cases} \qquad\text{or} \qquad
C_0-> X_0 = \begin{cases}
\omega^Tx^{(N+1)} \leq{-1}\\
\omega^Tx^{(N+2)} \leq{-1}\\
 \quad  \quad \vdots \\
\omega^Tx^{(N+M)} \leq{-1}\\
\end{cases}
\end{align*}
$$
<br>

<br>

__Matrix Form__

<br>

$$\begin{align*} C_1-> X_1 = \begin{bmatrix} \left(x^{(1)}\right)^T \\ \left(x^{(2)}\right)^T \\ \vdots \\ \left(x^{(N)}\right)^T\end{bmatrix} =
\begin{bmatrix} 1 & x_1^{(1)} & x_2^{(1)} \\ 1 & x_1^{(2)} & x_2^{(2)} \\
\vdots & \vdots & \vdots \\ 1 & x_1^{(N)} & x_2^{(N)} \\
\end{bmatrix} \qquad \qquad
C_1-> X_1 &= \begin{bmatrix} \left(x^{(1)}\right)^T \\ \left(x^{(2)}\right)^T \\ \vdots \\ \left(x^{(N)}\right)^T\end{bmatrix} =
\begin{bmatrix} x_1^{(1)} & x_2^{(1)} \\ x_1^{(2)} & x_2^{(2)} \\
\vdots & \vdots \\ x_1^{(N)} & x_2^{(N)} \\
\end{bmatrix}
\end{align*}$$

<br>

$$\begin{align*} C_0-> X_0 = \begin{bmatrix} \left(x^{(N+1)}\right)^T \\ \left(x^{(N+2)}\right)^T \\ \vdots \\ \left(x^{(N+M)}\right)^T\end{bmatrix} =
\begin{bmatrix} 1 & x_1^{(N+1)} & x_2^{(N+1)} \\ 1 & x_1^{(N+2)} & x_2^{(N+2)}\\
\vdots & \vdots & \vdots \\ 1 & x_1^{(N+M)} & x_2^{(N+M)} \\
\end{bmatrix} \qquad \qquad
C_0-> X_0 = \begin{bmatrix} \left(x^{(N+1)}\right)^T \\ \left(x^{(N+2)}\right)^T \\ \vdots \\ \left(x^{(N+M)}\right)^T\end{bmatrix} =
\begin{bmatrix} x_1^{(N+1)} & x_2^{(N+1)} \\ x_1^{(N+2)} & x_2^{(N+2)} \\
\vdots & \vdots \\ x_1^{(N+M)} & x_2^{(N+M)} \\
\end{bmatrix}
\end{align*}
$$

<br>

__Optimization in Form 1 (Standard Form with Separate Bias Term)__

<br>

$$\begin{align*}
\text{제약조건(Constrains)} \quad
\omega_0 + X_1\omega \geq 1 \\
\omega_0 + X_0\omega \leq -1
\end{align*}$$

<br>


```python
# CVXPY

import cvxpy as cvx

X1 = np.hstack([x1[C1], x2[C1]])
X0 = np.hstack([x1[C0], x2[C0]])

X1 = np.asmatrix(X1)
X0 = np.asmatrix(X0)

N = X1.shape[0]
M = X0.shape[0]
```

```python
M
```

```python
w0 = cvx.Variable([1,1])
w = cvx.Variable([2,1])

obj = cvx.Minimize(1)
const = [w0 + X1@w >= 1, w0 + X0@w <= -1]
prob = cvx.Problem(obj, const).solve()

w0 = w0.value
w = w.value
```

```python
w
```

```python
xp = np.linspace(0,9,100).reshape(-1,1)
yp = - w[0,0]/w[1,0]*xp - w0/w[1,0]

ypt = -2/3*xp + 6

plt.figure(figsize = (6, 4))
plt.plot(x1[C1], x2[C1], 'go', label = 'C1')
plt.plot(x1[C0], x2[C0], 'bo', label = 'C0')
plt.plot(xp, ypt, 'k', alpha = 0.3, label = 'True')
plt.plot(xp, ypt-1, '--k', alpha = 0.3)
plt.plot(xp, ypt+1, '--k', alpha = 0.3)
plt.plot(xp, yp, 'r', linewidth = 3, label = 'Attempt 1')
plt.title('Linearly and Strictly Separable Classes')
plt.xlabel(r'$x_1$')
plt.ylabel(r'$x_2$')
plt.legend(loc = 3)
plt.xlim([0, 9])
plt.ylim([0, 6])
plt.show()
```

\end{align*}$$

## 3.3. 개선된 최적화(Optimization) --> Soft Margin SVM

__제약조건을 완화(Relaxing the Constraints)__

- 선형 분류가 되지 않는 data(outlier)가 있어 w값을 얻을 수 없는 경우를 margin violation이라 함.
- 새로운 가변수(slack variables)를 도입하여 초기 제약조건을 완화시킴
- 가변수(slack variables) : $u$ and $v$
  - $u_i$ : Class 1에서의 margin violation 데이터
  - $v_i$ : Class 0에서의 margin violation 데이터

- $u_i \geq 0$ and $v_i \geq 0$

<br>

__선형 분류가 되지 않는 데이터에 대한 최적화(Optimization)__

- 가변수(slack variables)를 적용하면 margin violations을 최소화하는 최적화된 hyperplane이 존재하여 w값을 얻을 수 있음

<br>

$$\begin{align*} \text{minimize} \quad & \sum\limits_{i=1}^{N}u_i + \sum\limits_{i=1}^{M}\upsilon_i \\ \\
\text{제약조건 :} \quad
& \begin{cases}
\omega^Tx^{(1)} \geq 1-u_1\\
\omega^Tx^{(2)} \geq 1-u_2\\
 \quad  \quad \vdots \\
\omega^Tx^{(N)} \geq 1-u_N\\
\end{cases} \\\\
& \begin{cases}
\omega^Tx^{(N+1)} \leq {-(1-\upsilon_1)}\\
\omega^Tx^{(N+2)} \leq {-(1-\upsilon_2)}\\
 \quad  \quad \vdots \\
\omega^Tx^{(N+M)} \leq {-(1-\upsilon_M)}\\
\end{cases} \\\\
& \begin{cases}
u \geq 0\\
v \geq 0\\
\end{cases}
\end{align*}$$

<br>

__Matrix Form__

<br>

$$\begin{align*} C_1-> X_1 = \begin{bmatrix} \left(x^{(1)}\right)^T \\ \left(x^{(2)}\right)^T \\ \vdots \\ \left(x^{(N)}\right)^T\end{bmatrix} =
\begin{bmatrix} 1 & x_1^{(1)} & x_2^{(1)} \\ 1 & x_1^{(2)} & x_2^{(2)} \\
\vdots & \vdots & \vdots \\ 1 & x_1^{(N)} & x_2^{(N)} \\
\end{bmatrix} \qquad \qquad
u = \begin{bmatrix} u_1\\u_2\\ \vdots\\u_N\end{bmatrix}
\end{align*}$$

<br>

$$\begin{align*} C_0-> X_0 = \begin{bmatrix} \left(x^{(N+1)}\right)^T \\ \left(x^{(N+2)}\right)^T \\ \vdots \\ \left(x^{(N+M)}\right)^T\end{bmatrix} =
\begin{bmatrix} 1 & x_1^{(N+1)} & x_2^{(N+1)} \\ 1 & x_1^{(N+2)} & x_2^{(N+2)}\\
\vdots & \vdots & \vdots \\ 1 & x_1^{(N+M)} & x_2^{(N+M)} \\
\end{bmatrix} \qquad \qquad
v = \begin{bmatrix} v_1\\v_2\\ \vdots\\v_M\end{bmatrix}
\end{align*}$$

<br>

<br>

$$\begin{align*} \text{minimize} \quad  & 1^Tu + 1^T\upsilon \\
\text{제약조건 :} \quad
& X_1\omega \geq 1-u \\
& X_0\omega \leq -(1-\upsilon) \\
& u \geq 0 \\
& \upsilon \geq 0
\end{align*}$$

<br>

```python
w = cvx.Variable([3,1])
u = cvx.Variable([N,1])
v = cvx.Variable([M+1,1])

obj = cvx.Minimize(np.ones((1,N))@u + np.ones((1,M+1))@v)
const = [X1@w >= 1-u, X0@w <= -(1-v), u >= 0, v >= 0 ]
prob = cvx.Problem(obj, const).solve()

w = w.value
```

```python
xp = np.linspace(0,9,100).reshape(-1,1)
yp = - w[1,0]/w[2,0]*xp - w[0,0]/w[2,0]

ypt = -2/3*xp + 6

plt.figure(figsize = (6, 4))
plt.plot(X1[:,1], X1[:,2], 'go', label = 'C1')
plt.plot(X0[:,1], X0[:,2], 'bo', label = 'C0')
plt.plot(xp, ypt, 'k', alpha = 0.3, label = 'True')
plt.plot(xp, ypt-1, '--k', alpha = 0.3)
plt.plot(xp, ypt+1, '--k', alpha = 0.3)
plt.plot(xp, yp, 'r', linewidth = 3, label = 'Attempt 2')
plt.plot(xp, yp-1/w[2,0], '--g')
plt.plot(xp, yp+1/w[2,0], '--g')
plt.title('When Outliers Exist')
plt.xlabel(r'$x_1$')
plt.ylabel(r'$x_2$')
plt.legend(loc = 3)
plt.xlim([0, 9])
plt.ylim([0, 6])
plt.show()
```

__Further Improvement: Large Margin__

- 마진(Margin)의 폭을 최대화해서 완전한 최적화 구현

## 3.4. Maximize Margin --> Support Vector Machine(SVM)

- 마진(margin) : 두 직선간의 거리
<br>

$$\text{margin} = \frac{2}{\lVert \omega \rVert _2}$$
<br>
- 결정선(decision line)으로부터 가장 가까운 데이터의 거리(마진)를 최대화
<br><br>
    $$ \text{maximize {minimum distance}} $$
<br>
- 마진을 최대화 = $\lVert \omega \rVert_2$ 을 최소화
<br>
$$\begin{align*} \text{minimize} \quad
& \lVert \omega \rVert_2 + \gamma(1^Tu + 1^T\upsilon) \\
\text{제약조건 :} \quad
& X_1\omega \geq 1-u \\
& X_0\omega \leq -(1-\upsilon) \\
& u \geq 0 \\
& \upsilon \geq 0
\end{align*}$$
<br>
- gamma ($\gamma$) : 두 항목을 조절하는 하이퍼 파라미터
    - 마진을 최대화(Bigger margin given robustness to outliers) = $\lVert \omega \rVert_2$ 를 최소화
    
    - 에러를 최소화(Hyperplane that has few (or no) errors) --> $\\gamma(1^Tu + 1^T\upsilon)$

```python
gamma = 2

w = cvx.Variable([3,1])
u = cvx.Variable([N,1])
v = cvx.Variable([M+1,1])

obj = cvx.Minimize(cvx.norm(w,2) + gamma*(np.ones((1,N))@u + np.ones((1,M+1))@v))
const = [X1@w >= 1-u, X0@w <= -(1-v), u >= 0, v >= 0 ]
prob = cvx.Problem(obj, const).solve()

w = w.value
```

```python
xp = np.linspace(0,9,100).reshape(-1,1)
yp = - w[1,0]/w[2,0]*xp - w[0,0]/w[2,0]

ypt = -2/3*xp + 6

plt.figure(figsize = (6, 4))
plt.plot(X1[:,1], X1[:,2], 'go', label = 'C1')
plt.plot(X0[:,1], X0[:,2], 'bo', label = 'C0')
plt.plot(xp, ypt, 'k', alpha = 0.3, label = 'True')
plt.plot(xp, ypt-1, '--k', alpha = 0.3)
plt.plot(xp, ypt+1, '--k', alpha = 0.3)
plt.plot(xp, yp, 'r', linewidth = 3, label = 'Decision boundry')
plt.plot(xp, yp-1/w[2,0], '--g')
plt.plot(xp, yp+1/w[2,0], '--g')
plt.title('When Outliers Exist')
plt.xlabel(r'$x_1$')
plt.ylabel(r'$x_2$')
plt.legend(loc = 3)
plt.xlim([0, 9])
plt.ylim([0, 6])
plt.show()
```

## 3.5. Support Vector Machine
- compact form : $u$와 $v$를 하나의 문자 $\xi$로 표현
<br>
$$ \ \xi= \begin{bmatrix} u \\v \ \end{bmatrix}$$
<br>
$$
\begin{align*}
\omega^T x_n  &\geq 1 \;\text{for }\; y_n = +1 \\
\omega^T x_n  &\leq -1 \;\text{for }\; y_n = -1
\end{align*} \quad
\Longleftrightarrow \quad y_n \cdot \left( \omega^T x_n  \right) \geq 1
$$

<br>

$$\begin{align*} \text{minimize} \quad
& \lVert \omega \rVert_2 + \gamma(1^Tu + 1^T\upsilon) \\
\text{제약조건 :} \quad
& X_1\omega \geq 1-u \\
& X_0\omega \leq -(1-\upsilon) \\
& u \geq 0 \\
& \upsilon \geq 0
\end{align*}$$

$$\begin{align*} \quad
\Longleftrightarrow \quad \text{minimize} \quad
& \lVert \omega \rVert_2 + \gamma(1^T \xi) \\
\text{제약조건 :} \quad
& y_n \cdot \left( \omega^T x_n  \right) \geq 1 - \xi_n \\
& \xi \geq 0 \\
\end{align*}$$

<br>

```python
X = np.vstack([X1, X0])
y = np.vstack([np.ones([N,1]), -np.ones([M+1,1])])

m = N + M + 1

gamma = 2

w = cvx.Variable([3,1])
d = cvx.Variable([m,1])

obj = cvx.Minimize(cvx.norm(w,2) + gamma*(np.ones([1,m])@d))
const = [cvx.multiply(y, X@w) >= 1-d, d >= 0]
prob = cvx.Problem(obj, const).solve()

w = w.value
```

```python
xp = np.linspace(0,9,100).reshape(-1,1)
yp = - w[1,0]/w[2,0]*xp - w[0,0]/w[2,0]

ypt = -2/3*xp + 6

plt.figure(figsize = (6, 4))
plt.plot(X1[:,1], X1[:,2], 'go', label = 'C1')
plt.plot(X0[:,1], X0[:,2], 'bo', label = 'C0')
plt.plot(xp, ypt, 'k', alpha = 0.3, label = 'True')
plt.plot(xp, ypt-1, '--k', alpha = 0.3)
plt.plot(xp, ypt+1, '--k', alpha = 0.3)
plt.plot(xp, yp, 'r', linewidth = 3, label = 'Decision boundry')
plt.plot(xp, yp-1/w[2,0], '--g')
plt.plot(xp, yp+1/w[2,0], '--g')
plt.title('When Outliers Exist')
plt.xlabel(r'$x_1$')
plt.ylabel(r'$x_2$')
plt.legend(loc = 3)
plt.xlim([0, 9])
plt.ylim([0, 6])
plt.show()
```

#  4. Scikit-learn을 이용한 SVM 예제

- 이진분류(Binary classification)
  - $Class1 -> C1$ and $Class0 -> C0$

<br>

- 입력(Inputs or Features)
<br>
$$x = \begin{bmatrix}x_1 \\ x_2 \end{bmatrix}$$

<br>

- 분류경계선(hyperplane)
<br>

$$
\begin{align*}
\omega_0 + \omega_1x_1 + \omega_2x_2 &= 0 \\\\
\omega_0 + \begin{bmatrix}\omega_1 & \omega_2 \end{bmatrix}
\begin{bmatrix}x_1 \\ x_2 \end{bmatrix} &= 0\\\\
\omega_0 + \omega^Tx &= 0
\end{align*}
$$

<br>


$$\begin{align*} C_1-> X_1 &= \begin{bmatrix} \left(x^{(1)}\right)^T \\ \left(x^{(2)}\right)^T \\ \vdots \\ \left(x^{(N)}\right)^T\end{bmatrix} =
\begin{bmatrix} x_1^{(1)} & x_2^{(1)} \\ x_1^{(2)} & x_2^{(2)} \\
\vdots & \vdots \\ x_1^{(N)} & x_2^{(N)} \\
\end{bmatrix}
\end{align*}$$

<br>

$$\begin{align*} C_0-> X_0 = \begin{bmatrix} \left(x^{(N+1)}\right)^T \\ \left(x^{(N+2)}\right)^T \\ \vdots \\ \left(x^{(N+M)}\right)^T\end{bmatrix} =
\begin{bmatrix} x_1^{(N+1)} & x_2^{(N+1)} \\ x_1^{(N+2)} & x_2^{(N+2)} \\
\vdots & \vdots \\ x_1^{(N+M)} & x_2^{(N+M)} \\
\end{bmatrix}
\end{align*}
$$

<br>

```python
# Scikit-Learn

X1 = np.hstack([x1[C1], x2[C1]])
X0 = np.hstack([x1[C0], x2[C0]])
X = np.vstack([X1, X0])

N = X1.shape[0]
M = X0.shape[0]

y = np.vstack([np.ones([N,1]), np.zeros([M,1])])
```

```python
from sklearn import svm

clf = svm.SVC(kernel = 'linear')  # Support vector classifier
clf.fit(X, np.ravel(y))
```

# 5. Nonlinear Classification in SVM

```python
X1 = np.array([[-1.1,0],[-0.3,0.1],[-0.9,1],[0.8,0.4],[0.4,0.9],[0.3,-0.6],
               [-0.5,0.3],[-0.8,0.6],[-0.5,-0.5]])

X0 = np.array([[-1,-1.3], [-1.6,2.2],[0.9,-0.7],[1.6,0.5],[1.8,-1.1],[1.6,1.6],
               [-1.6,-1.7],[-1.4,1.8],[1.6,-0.9],[0,-1.6],[0.3,1.7],[-1.6,0],[-2.1,0.2]])

X1 = np.asmatrix(X1)
X0 = np.asmatrix(X0)

plt.figure(figsize = (6, 4))
plt.plot(X1[:,0], X1[:,1], 'ro', label = 'C1')
plt.plot(X0[:,0], X0[:,1], 'bo', label = 'C0')
plt.title('SVM for Nonlinear Data')
plt.xlabel(r'$x_1$')
plt.ylabel(r'$x_2$')
plt.legend(loc = 1)
plt.show()
```

<br>

$$
x = \begin{bmatrix} x_1 \\ x_2 \end{bmatrix} \quad \Longrightarrow \quad
z = \phi(x) = \begin{bmatrix} 1\\ x_1^2 \\ \sqrt{2}x_1 x_2 \\x_2^2 \end{bmatrix}$$

<br>

```python
N = X1.shape[0]
M = X0.shape[0]

X = np.vstack([X1, X0])
y = np.vstack([np.ones([N,1]), -np.ones([M,1])])

X = np.asmatrix(X)
y = np.asmatrix(y)

m = N + M
Z = np.hstack([np.ones([m,1]), np.square(X[:,0]), np.sqrt(2)*np.multiply(X[:,0],X[:,1]), np.square(X[:,1])])
```

```python
g = 10

w = cvx.Variable([4, 1])
d = cvx.Variable([m, 1])

obj = cvx.Minimize(cvx.norm(w, 2) + g*np.ones([1,m])@d)
const = [cvx.multiply(y, Z@w) >= 1-d, d>=0]
prob = cvx.Problem(obj, const).solve()

w = w.value
print(w)
```

```python
# to plot

[X1gr, X2gr] = np.meshgrid(np.arange(-3,3,0.1), np.arange(-3,3,0.1))

Xp = np.hstack([X1gr.reshape(-1,1), X2gr.reshape(-1,1)])
Xp = np.asmatrix(Xp)

m = Xp.shape[0]
Zp = np.hstack([np.ones([m,1]), np.square(Xp[:,0]), np.sqrt(2)*np.multiply(Xp[:,0], Xp[:,1]), np.square(Xp[:,1])])
q = Zp*w

B = []
for i in range(m):
    if q[i,0] > 0:
        B.append(Xp[i,:])

B = np.vstack(B)

plt.figure(figsize = (6, 4))
plt.plot(X1[:,0], X1[:,1], 'ro', label = 'C1')
plt.plot(X0[:,0], X0[:,1], 'bo', label = 'C0')
plt.plot(B[:,0], B[:,1], 'gs', markersize = 10, alpha = 0.1, label = 'SVM')
plt.title('SVM with Kernel')
plt.xlabel(r'$x_1$')
plt.ylabel(r'$x_2$')
plt.legend(loc = 1)
plt.show()
```
